```python
# import required libraries :
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# load data :
df1 = pd.read_csv("D:/LOAN APP PROJECT/data/cleaned_loan_approval_dataset.csv")
```

```python
# copy Data : for next process :
df2 = df1.copy()
```

## EXPLORATORY DATA ANALYSIS : 

### STEP 1 : HIGH LEVEL UNDERSTANDING : 

```python
# normalize columns -> snake_case (safe to use afterwards)
df2.columns = df2.columns.str.strip().str.lower().str.replace(' ', '_')
print("New column names:")
print(list(df2.columns))
print("\nDataset shape:", df2.shape)
df2.head()
```

```python
# Info & dtypes
df2.info()        # columns, dtypes, non-null counts
print("\nRow × Col:", df2.shape)
```

```python
#  Missing values summary ---
null_count = df2.isnull().sum()
null_pct = (null_count/len(df2)*100).round(2)
nulls = pd.DataFrame({'null_count':null_count, 'null_percent':null_pct})
nulls = nulls.sort_values('null_count', ascending=False)
nulls
```

```python
# Unique values per column (cardinality) : 
card = df2.nunique().sort_values(ascending=False).to_frame('unique_values')
card
```

```python
# Target distribution
target = 'loan_status'
print("Unique values in target:", df2[target].unique())
print("\nCounts:")
print(df2[target].value_counts(dropna=False))

print("\nPercent:")
print((df2[target].value_counts(normalize=True, dropna=False)*100).round(2))
```

```python
# numeric summary : 
num_cols = df2.select_dtypes(include=[np.number]).columns.tolist()
if 'loan_id' in num_cols :
    num_cols.remove("loan_id")
num_summary = df2[num_cols].describe().T.round(2)
num_summary
```

```python
#  Categorical summary (top values)
cat_cols = df2.select_dtypes(include=['object', 'category']).columns.tolist()
cat_cols = [c for c in cat_cols if c != 'loan_status']  # exclude target
cat_summary = {}
for c in cat_cols:
    top = df2[c].value_counts().head(6)
    cat_summary[c] = top
    print(f"\n--- {c} ---")
    print(top)
```

```python
#  Sanity checks ---
# 1) loan_id uniqueness
if 'loan_id' in df2.columns:
    print("loan_id unique?", df2['loan_id'].is_unique)
    print("Any duplicates (count):", df2['loan_id'].duplicated().sum())

# 2) Negative checks for financial columns
money_cols = [c for c in ['residential_assets_value','residential_assets_value_clean',
                          'commercial_assets_value','luxury_assets_value','bank_asset_value',
                          'income_annum','loan_amount'] if c in df2.columns]
neg_counts = (df2[money_cols] < 0).sum()
print("\nNegative value counts (should be 0):")
print(neg_counts)

# 3) CIBIL sanity
if 'cibil_score' in df2.columns:
    out_of_range = df2[(df2['cibil_score'] < 300) | (df2['cibil_score'] > 900)]
    print("\nCIBIL out-of-range rows:", len(out_of_range))
```

```python
# Income_bin & Loan_bin distribution ---
for c in ['income_bin','loan_bin']:
    if c in df2.columns:
        print(f"\nDistribution for {c}:")
        print(df2[c].value_counts(normalize=True).round(4) * 100)
```

```python
#  Compare original vs cleaned residential assets distribution ---
if 'residential_assets_value' in df2.columns and 'residential_assets_value_clean' in df2.columns:
    plt.figure(figsize=(10,4))
    sns.histplot(df2['residential_assets_value'].clip(lower=0), label='original (clipped)', stat='density', kde=False, alpha=0.4, bins=50)
    sns.histplot(df2['residential_assets_value_clean'].clip(lower=0), label='cleaned', stat='density', kde=False, alpha=0.4, bins=50)
    plt.legend(); plt.title("Residential assets: original vs cleaned (overlay)")
    plt.show()
```

## STEP 2 : LOW LEVEL UNDERSTANDING : 

### STEP 2.1 : UNIVARIATE ANALYSIS : 

```python
# 1) Univariate - Numerical distributions : 
num_cols = df2.select_dtypes(include=[np.number]).columns.tolist()
if 'loan_id' in num_cols :
    num_cols.remove('loan_id')
n = len(num_cols)
rows = math.ceil(n/3)
plt.figure(figsize=(16,rows*4))
for i, col in enumerate(num_cols, 1) :
    plt.subplot(rows, 3, i)
    sns.histplot(x=df2[col], kde=True, bins=10, color='darkblue')
    plt.title(f"{col} distribution")

plt.tight_layout()
plt.show()
```

```python
# 2) Univariate - categories distributions : 
cat_cols = df2.select_dtypes(include=['object','category']).columns.to_list()
n = len(cat_cols)
rows = math.ceil(n/2)
plt.figure(figsize=(15,20))
for i,col in enumerate(cat_cols, 1) :
    plt.subplot(rows, 2, i)
    sns.countplot(x=df2[col], order=df2[col].value_counts().index)
    
    plt.title(f"{col} Frequency")
plt.tight_layout()
plt.show()
```

```python
# 3) Target Variable 
sns.countplot(data=df2, x='loan_status', palette='Set1')
plt.title('Loan Status Distribution')
plt.show()
```

### STEP 2.2 BIVARIATE ANALYSIS : 

```python
# Numerical vs Target (laon_status)
df2.groupby('loan_status')[num_cols].mean().round(2)
```

```python
# 1) Numerical vs Target (lona_status) : 

num_cols = df2.select_dtypes(include=[np.number]).columns.to_list()
if 'loan_id' in num_cols : 
    num_cols.remove('loan_id')

# display box plot : for outlier detection 
n = len(num_cols)
rows = math.ceil(n/3)
plt.figure(figsize=(16, rows*4))
for i, col in enumerate(num_cols, 1) :
    plt.subplot(rows, 3, i)
    sns.boxplot(data=df2, x='loan_status', y=col)
    plt.title(f"{col} vs loan_status")
    
plt.suptitle("Numerical Features vs Loan Status (Boxplots)", fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# 2) Categorical Vs Target(loan_status) fetures : 
cat_cols = ['education','self_employed']

n = len(cat_cols)
rows = math.ceil(n / 2)

fig, axes = plt.subplots(rows, 2, figsize=(14, rows * 4))  # shared y for easier comparison
axes = axes.flatten()  # Flatten in case of single row

for i, col in enumerate(cat_cols):
    cross = pd.crosstab(df2[col], df2['loan_status'], normalize='index') * 100
    cross.plot(kind='bar', stacked=True, colormap="Set1", ax=axes[i])

    axes[i].set_title(f"{col} vs Loan_status (%)")
    axes[i].set_ylabel("Percentage")
    axes[i].set_xlabel(col)
    axes[i].legend(title="Loan Status")

# Hide any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle("Categorical Features vs Loan Status (%)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for suptitle
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Define categorical columns
cat_cols = ['education', 'self_employed']

for col in cat_cols:
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(14, 4))
    fig.suptitle(f"{col} vs Loan Status", fontsize=14, fontweight='bold')

    # 1️⃣ Cross Table
    cross_tab = pd.crosstab(df2[col], df2['loan_status'])
    axes[0].axis('off')
    axes[0].set_title("Cross Table", fontsize=12)
    table = axes[0].table(
        cellText=cross_tab.values,
        colLabels=cross_tab.columns,
        rowLabels=cross_tab.index,
        loc='center',
        cellLoc='center'
    )
    table.scale(1.2, 1.2)

    # 2️⃣ Stacked Bar Chart
    cross_tab.plot(
        kind="bar",
        stacked=True,
        ax=axes[1],
        colormap='Set2'
    )
    axes[1].set_title("Stacked Bar Chart", fontsize=11)
    axes[1].set_xlabel(col)
    axes[1].set_ylabel("Count")

    # 3️⃣ Grouped Bar Chart
    cross_tab.plot(
        kind="bar",
        ax=axes[2],
        colormap='Set1'
    )
    axes[2].set_title("Grouped Bar Chart", fontsize=11)
    axes[2].set_xlabel(col)
    axes[2].set_ylabel("Count")

    # Adjust layout for neat spacing
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
# 3) numerical vs numerical features analysis : 
# --- Correlation heatmap (numeric) ---
plt.figure(figsize=(12,4))
sns.heatmap(df2[num_cols].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation heatmap")
plt.show()
```

### STPE 2.3 MULTIVARIATE ANALYSIS : 

```python
# 1) Pairplot (sample of important features) ---
sample_cols = ['income_annum','loan_amount','cibil_score','bank_asset_value','loan_status']
sns.pairplot(df2[sample_cols], hue='loan_status', diag_kind='kde', corner=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Income vs Loan Status by Education
g = sns.catplot(
    data=df2,
    x='loan_status',
    y='income_annum',
    hue='education',
    kind='bar',
    height=4.5,
    aspect=1.3,
    palette='Set1'
)

# Add average labels on bars
for ax in g.axes.flat:
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f', fontsize=8, label_type='edge', padding=2)

# Remove default legend
g._legend.remove()

# Add custom legend (only one)
plt.legend(
    title='Education',
    loc='upper right',
    bbox_to_anchor=(1.18, 1),
    fontsize=9,
    title_fontsize=10,
    frameon=False
)

# Titles and labels
g.fig.suptitle("Income vs Loan Status by Education", fontsize=13, fontweight='bold')
g.set_axis_labels("Loan Status", "Average Annual Income")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()


# 2️⃣ Loan Amount vs Loan Status by Self Employment
g2 = sns.catplot(
    data=df2,
    x='loan_status',
    y='loan_amount',
    hue='self_employed',
    kind='bar',
    height=4.5,
    aspect=1.3,
    palette='Set1'
)

# Add average labels on bars
for ax in g2.axes.flat:
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f', fontsize=8, label_type='edge', padding=2)

# Remove default legend
g2._legend.remove()

# Add custom legend (only one)
plt.legend(
    title='Self Employed',
    loc='upper right',
    bbox_to_anchor=(1.18, 1),
    fontsize=9,
    title_fontsize=10,
    frameon=False
)

# Titles and labels
g2.fig.suptitle("Loan Amount vs Loan Status by Self Employment", fontsize=13, fontweight='bold')
g2.set_axis_labels("Loan Status", "Average Loan Amount")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
